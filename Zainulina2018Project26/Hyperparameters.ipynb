```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn-whitegrid')

import warnings
warnings.filterwarnings("ignore")

plt.rcParams.update({'font.size': 17})
PATH = 'grid_results'
```

```python
def graph(path, name):
#     arr = arr.tolist()
    x = [0.0001, 0.001, 0.01, 0.1]
    plt.figure(figsize = (16, 6))
    
    for i, C in enumerate([1, 10]):
        for folder in ['dataset_1', 'dataset_2', 'dataset_3']:
            arr = np.load(path+'/'+folder+'/file_'+name+'.npy').tolist()
            test_mn = np.abs(arr['mean_test_score'][4*(C>5):(8-4*(C<5))])
            test_std = np.abs(arr['std_test_score'][4*(C>5):(8-4*(C<5))])
            plt.subplot(1,2,1+i)
            plt.plot(x, test_mn, label=folder)
            plt.fill_between(x, test_mn-test_std, test_mn+test_std, alpha=0.3)
            print(folder,': C=',C, 'gamma=', x[np.argmin(test_mn)], 'MSE=', np.amin(test_mn))
#             if folder == 'dataset_3':
#                 for num in np.around(test_mn, decimals=5):
#                     print ('& $'+str(num)+'$'),
        plt.xlabel('gamma')
        plt.ylabel('MSE, $m^2/s^2$')
        plt.title('C='+str(C))
        plt.legend()
        if C == 10:
            plt.savefig('graphs/'+name+'_C='+str(C)+'.pdf')
```
