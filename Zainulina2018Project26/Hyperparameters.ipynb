```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn-whitegrid')

import warnings
warnings.filterwarnings("ignore")
```

```python
def graph(arr, name):
    arr = arr.tolist()
    x = [0.0005, 0.0008, 0.001, 0.005, 0.01, 0.013]
    plt.figure(figsize = (16, 6))
    for i, C in enumerate([1, 10]):
        train_mn = np.abs(arr['mean_train_score'][6*(C>5):(12-6*(C<5))])
        train_std = np.abs(arr['std_train_score'][6*(C>5):(12-6*(C<5))])
        test_mn = np.abs(arr['mean_test_score'][6*(C>5):(12-6*(C<5))])
        test_std = np.abs(arr['std_test_score'][6*(C>5):(12-6*(C<5))])
        plt.subplot(1,2,1+i)
        plt.plot(x, train_mn, label='train')
        plt.fill_between(x, train_mn-train_std, train_mn+train_std, alpha=0.3)
        plt.plot(x, train_mn, label='test')
        plt.fill_between(x, train_mn-train_std, train_mn+train_std, alpha=0.3)
        plt.xlabel('epsilon')
        plt.ylabel('MSE')
        plt.title('Dan_'+name+', C='+str(C))
```

```python
bag0 = np.load('file_bag_chn0.npy')
graph(bag0, 'bag2_chn0')
```
